建物型態 :  {住宅大樓(11層含以上有電梯),公寓(5樓含以下無電梯),套房(1房1廳1衛),店面(店鋪),華廈(10層含以下有電梯),辦公商業大樓,透天厝}

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras

tf.compat.v1.reset_default_graph()

df = pd.read_csv('士林區.csv')
df = np.array(df)

new_model = keras.models.load_model('Sequential_3.h5')

pred = new_model.predict(df)

y = np.expm1(pred)

y = y.reshape((1,7))[0]
y1 = round(y[0],1)
y2 = round(y[1],1)
y3 = round(y[2],1)
y4 = round(y[3],1)
y5 = round(y[4],1)
y6 = round(y[5],1)
y7 = round(y[6],1)

print('='*100)
print('住宅大樓(11層含以上有電梯):',y1)
print('公寓(5樓含以下無電梯):',y2)
print('套房(1房1廳1衛):',y3)
print('店面(店鋪):',y4)
print('華廈(10層含以下有電梯):',y5)
print( '辦公商業大樓:',y6)
print('透天厝:',y7)
```
